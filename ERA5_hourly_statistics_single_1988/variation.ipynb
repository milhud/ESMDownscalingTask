## Variation, Correlation

Let's now make a scatter plot comparing precipitation with respect to temperature.  Let's begin by loading the needed data.

```python
import xarray as xarr

filename = "./data/clipped_hourly_single_1988.nc"
data = xarr.open_dataset(filename)

precipitation_data = data["tp"].mean(dim=["latitude","longitude"])
temp_data = data["t2m"].mean(dim=["latitude", "longitude"])
```

Using the following [reference](https://docs.xarray.dev/en/stable/generated/xarray.DataArray.polyfit.html), we can now look to define a scatter plot.

```python
import matplotlib.pyplot as plot

plot.figure(figsize=(12, 8))


# plot the scatter, fit line
plot.scatter(temp_data.values, precipitation_data.values, alpha=0.7)

# labels, save, show
plot.xlabel("Temperature (K)")
plot.ylabel("Precipitation (m/h)")
plot.title("Temperature vs Precipitation")
plot.savefig("./plots/Temperature_Precipitation_Correlation.png")
plot.show()
```
